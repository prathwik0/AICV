```python
img1 = cv2.imread("../imagedata/1bit1.png")
img2 = cv2.imread("../imagedata/1bit2.png")
```

```python
# Image Opeation AND
bitwiseAnd = cv2.bitwise_and(
    img1, img2, mask=None
)  # It creates a region on which the operation should be applied.
plt.imshow(bitwiseAnd)
# display("AND", bitwiseAnd)
```

```python
# Image Operation OR
bitwiseOr = cv2.bitwise_or(img1, img2)
plt.imshow(bitwiseOr)
# display("OR", bitwiseOr)
```

```python
# Image Operation Not
bitswiseNot = cv2.bitwise_not(img2)
plt.imshow(bitswiseNot)
# display("NOT", bitswiseNot)
```

```python
# Image Operation XOR
bitwiseXor = cv2.bitwise_xor(img1, img2)
plt.imshow(bitwiseXor)
# display("XOR", bitwiseXor)
```
